```python
# Step 1: NumPy Operations

import numpy as np
import time

# Create a NumPy array of numbers from 1 to 10
arr = np.arange(1, 11)
print("Original array:", arr)

# Elementwise operations
arr_squared = arr ** 2
arr_cubed = arr ** 3

print("Squared:", arr_squared)
print("Cubed:", arr_cubed)

# Compare loop vs vectorized execution
# Using loop
start = time.time()
squared_loop = []
for x in arr:
    squared_loop.append(x ** 2)
end = time.time()
print("Loop squared:", squared_loop)
print("Loop time:", end - start)

# Using vectorized NumPy
start = time.time()
squared_vector = arr ** 2
end = time.time()
print("Vectorized squared:", squared_vector)
print("Vectorized time:", end - start)
```

```python
# Step 0: Generate sample stock data -> ../data/starter_data.csv
import numpy as np
import pandas as pd
from pathlib import Path

np.random.seed(42)

# Two tickers over 60 business days
dates = pd.date_range(start="2024-01-01", periods=60, freq="B")
tickers = ["AAPL", "MSFT"]

rows = []
for t in tickers:
    base = 150 if t == "AAPL" else 320  # start prices differ a bit
    # simple random walk for price
    price = base + np.cumsum(np.random.normal(0, 1, len(dates)))
    high = price + np.abs(np.random.normal(0, 1, len(dates))) * 1.5
    low = price - np.abs(np.random.normal(0, 1, len(dates))) * 1.5
    openp = price + np.random.normal(0, 0.5, len(dates))
    close = price + np.random.normal(0, 0.5, len(dates))
    adj_close = close  # for this synthetic data, keep same as close
    volume = np.random.randint(1_000_000, 5_000_000, len(dates))

    for d, o, h, l, c, ac, v in zip(dates, openp, high, low, close, adj_close, volume):
        rows.append(
            (d.date(), t, round(o, 2), round(h, 2), round(l, 2), round(c, 2), round(ac, 2), v)
        )

df = pd.DataFrame(
    rows, columns=["Date", "Ticker", "Open", "High", "Low", "Close", "Adj Close", "Volume"]
)

# Ensure output folder exists and save
out_dir = Path("../data")
out_dir.mkdir(parents=True, exist_ok=True)
out_path = out_dir / "starter_data.csv"
df.to_csv(out_path, index=False)

print("✅ Sample dataset created at:", out_path)
df.head()
```

```python
# Step 2: Import pandas
import pandas as pd

# Step 2a: Create synthetic stock data with Date column
data = {
    "Date": ["2024-01-01","2024-01-02","2024-01-03","2024-01-04","2024-01-05"],
    "Ticker": ["AAPL"]*5,
    "Open": [150.81, 149.93, 150.47, 152.77, 152.18],
    "High": [151.22, 150.64, 152.67, 154.32, 153.51],
    "Low": [149.31, 148.99, 148.90, 150.43, 151.41],
    "Close": [150.10, 150.30, 151.26, 152.96, 151.69],
    "Adj Close": [150.10, 150.30, 151.26, 152.96, 151.69],
    "Volume": [4369720, 1516771, 3928009, 3660196, 3640328]
}

df = pd.DataFrame(data)

# Step 2b: Save to CSV
df.to_csv("synthetic_stock_data.csv", index=False)

# Step 2c: Reload the CSV (to mimic homework step of loading a dataset)
df = pd.read_csv("synthetic_stock_data.csv")

# Step 2d: Inspect dataset
print(df.head())
print(df.info())
print("Columns:", df.columns.tolist())

# Step 2e: Convert 'Date' to datetime and set as index
df["Date"] = pd.to_datetime(df["Date"])
df.set_index("Date", inplace=True)

print("\nAfter setting Date as index:")
print(df.head())
```

```python
# Step 3: Summary Statistics

# Numeric summary
numeric_summary = df.describe()
print("Numeric Summary:\n", numeric_summary)

# Group by Ticker (category) and aggregate
grouped_summary = df.groupby('Ticker').agg({
    'Open': ['mean', 'min', 'max'],
    'Close': ['mean', 'min', 'max'],
    'Volume': ['sum', 'mean']
})
print("\nGrouped Summary by Ticker:\n", grouped_summary)
```

```python
# Step 4: Save Outputs

# Save numeric summary to CSV
numeric_summary.to_csv('../data/processed/summary_numeric.csv')

# Save grouped summary to CSV
grouped_summary.to_csv('../data/processed/summary_grouped.csv')

print("Summary files saved in data/processed folder.")
```

```python
import pandas as pd

# Load the CSV you already have (replace path if different)
df = pd.read_csv("../data/starter_data.csv")  # or the original stock CSV

# Convert Date column to datetime and set as index
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)

df.head()
```

```python
df.to_csv("../data/processed/summary.csv")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
df['Close'].plot(title='Stock Closing Prices')
plt.xlabel('Date')
plt.ylabel('Close Price')

plt.savefig('../data/processed/closing_prices_plot.png')
plt.show()
```

```python
# Calculate daily returns
df['Daily_Return'] = df['Close'].pct_change()  # Percentage change from previous day

# Check first few rows
df.head()
```

```python
plt.figure(figsize=(10,5))
df['Daily_Return'].plot(title='Daily Returns')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.axhline(0, color='red', linestyle='--')  # reference line at 0
plt.show()
```

```python
# Summary statistics
daily_return_summary = df['Daily_Return'].describe()
print("Daily Returns Summary:")
print(daily_return_summary)
```

```python
# Save summary to a CSV file
daily_return_summary.to_csv('../data/processed/daily_return_summary.csv')
```

```python
import matplotlib.pyplot as plt

# Plot histogram of daily returns
plt.figure(figsize=(10,5))
df['Daily_Return'].hist(bins=50, alpha=0.7)
plt.title('Histogram of Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')

# Save the plot
plt.savefig('../data/processed/daily_return_histogram.png')
plt.show()
```

```python
print(df['Daily_Return'].dtype)
```

```python
# Clip returns at 1st and 99th percentile to reduce outlier effect
lower = df['Daily_Return'].quantile(0.01)
upper = df['Daily_Return'].quantile(0.99)
df_clipped = df['Daily_Return'].clip(lower, upper)
```

```python
plt.figure(figsize=(10,5))
df_clipped.hist(bins=50, alpha=0.7)
plt.title('Histogram of Daily Returns (Clipped)')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.show()
```

```python
# Step 8: Cumulative Returns
df['Cumulative_Return'] = (1 + df['Daily_Return']).cumprod()

# Plot cumulative returns
plt.figure(figsize=(10,5))
df['Cumulative_Return'].plot(title='Cumulative Returns Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.grid(True)
plt.show()
```

```python
# Step 9: Moving Averages
# 5-day and 20-day moving averages
df['MA_5'] = df['Close'].rolling(window=5).mean()
df['MA_20'] = df['Close'].rolling(window=20).mean()

# Plot closing prices with moving averages
plt.figure(figsize=(12,6))
df['Close'].plot(label='Close Price')
df['MA_5'].plot(label='5-Day MA')
df['MA_20'].plot(label='20-Day MA')
plt.title('Stock Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 10: Save processed data
df.to_csv("../data/processed/stock_data_processed.csv", index=True)

# Save the moving averages plot
plt.figure(figsize=(12,6))
df['Close'].plot(label='Close Price')
df['MA_5'].plot(label='5-Day MA')
df['MA_20'].plot(label='20-Day MA')
plt.title('Stock Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.savefig("../data/processed/moving_averages_plot.png")
plt.close()  # Close the plot to avoid overlap with future plots

# Optionally, save daily returns histogram
plt.figure(figsize=(10,5))
df['Daily_Return'].hist(bins=50)
plt.title('Histogram of Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.grid(False)
plt.savefig("../data/processed/daily_returns_histogram.png")
plt.close()
```

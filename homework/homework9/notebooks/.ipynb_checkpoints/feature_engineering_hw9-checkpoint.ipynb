```python
# Step 2: Create synthetic dataset for Homework 9
import pandas as pd
import numpy as np

np.random.seed(9)
n = 160

df = pd.DataFrame({
    'date': pd.date_range('2023-01-01', periods=n, freq='D'),
    'region': np.random.choice(['North','South','East','West'], size=n),
    'age': np.random.normal(40, 8, size=n).clip(22, 70).round(1),
    'income': np.random.lognormal(mean=10.6, sigma=0.3, size=n).round(2),
    'transactions': np.random.poisson(lam=3, size=n)
})

# Create target variable 'spend' as a function of income and transactions
base = df['income'] * 0.0015 + df['transactions']*18 + np.random.normal(0, 40, size=n)
df['spend'] = np.maximum(0, base).round(2)

# Introduce some missing values and outliers
df.loc[np.random.choice(df.index, 5, replace=False), 'income'] = np.nan
df.loc[np.random.choice(df.index, 3, replace=False), 'spend'] = np.nan
df.loc[np.random.choice(df.index, 2, replace=False), 'transactions'] = df['transactions'].max() + 12

# Preview the dataset
df.info()
df.head()
```

### Synthetic Dataset Creation
We created a synthetic dataset with 160 rows and the following variables:
- `date`: Daily timestamps
- `region`: Categorical region
- `age`: Numeric, customer age
- `income`: Numeric, customer income
- `transactions`: Numeric, number of transactions
- `spend`: Numeric, target variable reflecting spend
Some missing values and outliers were injected to simulate real-world conditions.


```python
# Step 3: Feature Engineering

# Feature 1: Spend-to-Income Ratio
df['spend_income_ratio'] = df['spend'] / df['income']

# Feature 2: Transactions per Age
df['transactions_per_age'] = df['transactions'] / df['age']

# Preview new features
df[['spend_income_ratio', 'transactions_per_age']].head()
```

### Feature Engineering Rationale

1. **spend_income_ratio**: Measures how much of a customer's income is spent. 
   - Helps capture proportionality of spending relative to income.
   - Can highlight high spenders relative to their earnings.

2. **transactions_per_age**: Normalizes transaction frequency by age.
   - Accounts for differences in activity levels across age groups.
   - Useful for understanding spending behavior relative to life stage.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation matrix
corr = df[['income','spend','transactions','spend_income_ratio','transactions_per_age']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# Scatterplots
sns.scatterplot(data=df, x='spend_income_ratio', y='spend', hue='region')
plt.title('Spend vs Spend-to-Income Ratio')
plt.show()

sns.scatterplot(data=df, x='transactions_per_age', y='spend')
plt.title('Spend vs Transactions per Age')
plt.show()
```

### Observations

1. **spend_income_ratio vs spend**: Higher ratio generally corresponds to higher spend, but some outliers exist.
2. **transactions_per_age vs spend**: Weak positive trend; age-normalized transaction activity slightly correlates with spend.
3. Correlation matrix shows moderate relationship between spend_income_ratio and spend, while transactions_per_age has weaker correlation.


## Top Insights

1. `spend_income_ratio` captures proportionality of spending to income and moderately correlates with spend.
2. `transactions_per_age` normalizes transactions by age; weak correlation with spend, but helps feature scaling.
3. Missing values and outliers from EDA should be addressed before modeling.

### Assumptions & Risks
- Assumes linear relationships in new features.
- Outliers in spend may skew ratio-based features.
- Age distribution impacts `transactions_per_age` metric.
